## Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

## Extracción de datos

```python
directorio = './datos'

reportes_homicidio = pd.read_csv(directorio + '/reportes-homicidio.csv')
```

## Número de incidentes por año

```python
plot.figure(figsize = (10, 5))
sns.barplot(x = 'Year', y = 'Incident', hue = 'Crime Solved', data = reportes_homicidio.groupby(['Year', 'Crime Solved']).sum().reset_index()).set_title('Incidencias por año')
plot.xticks(rotation = 90)
```

**Análisis:** Como se puede observar, el número de incidentes reportados ha ido en constante crecimiento desde 1985 hasta encontrar su punto más bajo entre los años de 1998 y 1999. A partir del nuevo siglo, los reportes de incidentes han aumentado considerablemente en comparación con las décadas pasadas, manteníendose el número de casos sin resolver más o menos de manera uniforme pero con una mayor cantidad de crímenes resueltos. Se puede apreciar además, que los años con mayor número de inicidentes resueltos fueron 2010 y 2006, quedando sin resolver, en los respectivos años, poco más de un tercio de los incidentes reportados.

## Número de incidentes por mes desde el 2005

```python
plot.figure(figsize = (20, 5))
sns.barplot(x = 'Month', y = 'Incident', hue = 'Year', data = reportes_homicidio[reportes_homicidio['Year'] >= 2005].groupby(['Month', 'Year']).sum().reset_index()).set_title('Incidencias por mes entre 2005 y 2014')
```

**Análisis:** En la gráfica anterior, se observa que a lo largo de 10 años, el número de incidentes reportados durante el mes de Junio entre los años 2005 y 2014 es prácticamente el mismo, correspondiendo dicho mes, a las vacaciones de verano escolares para los niños y jóvenes de Estados Unidos. Se puede observar también qu durante el mens de octubre de los años 2006 y 2010 se ha reportado una cantidad igual de incidentes, siendo una coincidencia no tan agradable. 

## Número de incidentes reportados por raza de los perpetradores

```python
plot.figure(figsize = (10, 5))
sns.barplot(x='Perpetrator Race', y='Incident', hue='Perpetrator Sex', data=reportes_homicidio)
```

**Análisis:** En la gráfica es evidente que el sexo de los culpables de homicidio es desconocido y corresponde justamente a la cantidad de reportes que no han podido ser resueltos. Se evidencia también, que la mayoría de los incidintes cometidos han sido provocados por personas caucásicas. Para el caso de perpetradores que son nativos, es mucho más sencillo conocer su sexo pues, para el caso de nativos de Alaska, el rango operacional de estos individuos es bastante limitado. 

## Estados con más de 10000 reportes emitidos durante 1980 a 2014

```python
estados = np.unique(reportes_homicidio['State'].get_values(), return_counts = True)

numero_incidencias_por_estado = estados[1][np.where(estados[1] > 10000)]
estados_mas_peligrosos = estados[0][np.where(estados[1] > 10000)]

plot.figure(figsize = (10, 10)) 
plot.pie(numero_incidencias_por_estado, 
         labels = estados_mas_peligrosos, 
         autopct = "%1.2f%%")
plot.title('Estados más peligrosos')
```

**Análisis:** En base al gráfico pastel, se pueden definir cuáles son los estados más peligrosos de Estados Unidos a través del número de reportes emitidos durante casi 40 años. Como se puede notar, el estado más peligroso, en base a dichos datos, es California, seguida de Nueva Your y Texas respectivamente. Otros estados que pueden considerarse como peligrosos son también Florida y Míchigan. 

## Ciudades más peligrosas del estado de California

```python
california = reportes_homicidio.loc[reportes_homicidio['State'] == 'California']['City']

ciudades_california = np.unique(california.get_values(), return_counts = True)

numero_incidencias_por_ciudad = ciudades_california[1][np.where(ciudades_california[1] > 1000)]
ciudades_mas_peligrosas = ciudades_california[0][np.where(ciudades_california[1] > 1000)]

plot.figure(figsize = (10, 10)) 
plot.pie(numero_incidencias_por_ciudad,
         labels = ciudades_mas_peligrosas,
         autopct = "%1.2f%%")
plot.title('Ciudades más peligrosas de California')
```

**Análisis:** En gráficas anteriores, se determinó que el estado más peligroso, en base a los reportes de homicidios emitidos durante el período de 1980 a 2014 y con esta información, se determina que la ciudad más peligrosa y sin dudar, es Los Ángeles. Por tanto, se puede concluir que la ciudad con mayor índice de homicidios es Los Ángeles. 

## Armas utilizadas en los homicidios reportados desde 1980 a 2014

```python
plot.figure(figsize = (15, 5)) 
sns.lineplot(x = "Year", y = "Incident", hue = "Weapon", data = reportes_homicidio)
```

**Análisis:** Por la variante mostrada de armas, se puede observar que a lo largo de los años, los incidentes registrados han sido provocados por objetos punzantes, drogas y armas de fuego. Cabe mencionar que durante finales de 1999 y mediados del 2003, la mayor cantidad de incidentes registrados fue a causa de drogas y con el pasar de los años, las causas fueron "normalizándose" y los incidentes que se registraron en mayor medida fueron con objetos punzantes y con armas principalmente.
