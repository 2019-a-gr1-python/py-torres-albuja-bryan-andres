## Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

## Extracción de datos

```python
directorio = './smashito/'

personajes = pd.read_csv(directorio + 'personajes.csv')
personajes_SSB4 = pd.read_csv(directorio + 'personajes-ssb4.csv')
```

## Crecimiento del número de personajes desde el primer juego hasta el último

Se definen variables para almacenar el numero de personajes (Fighters) para cada entrega de la franquicia

```python
personajes_smashito = pd.DataFrame(personajes['SSB']).count()
personajes_smashito_melee = pd.DataFrame(personajes['Melee']).count()
personajes_smashito_brawl = pd.DataFrame(personajes['Brawl']).count()
personajes_smashito_4 = pd.DataFrame(personajes['SSB4']).count()
personajes_smashito_ultimate = pd.DataFrame(personajes['Ult.']).count()
```

En base a las variables anteriores, se define una nueva variable que indicará el crecimiento porcentual del número de personajes desde el primer juego (SSB) hasta el último juego(Ult.)

```python
aumento_personajes = (personajes_smashito_ultimate[0] - personajes_smashito[0])*personajes['Fighter'].count()/100
```

En base a estos datos se obtiene:

```python
print(f"A lo largo de los años, el numero de personajes a aumentado de {personajes['SSB'].count()} a {personajes['Ult.'].count()} personajes. Aumentando estos en un {aumento_personajes}%")
```

## Número de personajes por entrega (juego)

En base a las variables antes definidas, se puede observar, de manera gráfica, el crecimiento del número de personajes por entrega:

```python
juegos = ['SSB', 'Melee', 'Brawl', 'SSB4', 'Ult.']
numero_personajes_por_juego = [personajes_smashito[0], personajes_smashito_melee[0], personajes_smashito_brawl[0], personajes_smashito_4[0], personajes_smashito_ultimate[0]]

plot.figure(1)
plot.bar(juegos, numero_personajes_por_juego)
plot.xticks(rotation=90)
plot.xlabel('Juegos')
plot.ylabel('No. de personajes por juego')
plot.title('Personajes por juego')
```

## Cantidad de personajes por universo

Dentro del dataset analizado, un universo se refiere al "mundo" al que pertenecen los múltiples personajes de Smash Bros. Por ejemplo, un universo es *_The Legend of Zelda_* pertenecinete a la franquicia de Nintendo. Al analizar los personajes por universo es indicar el número de personajes por cada "mundo" que ha sido incluido en los juegos de Smash Bros. a lo largo de los años.

```python
universos = personajes['Universe'].unique()
numero_personajes_por_universo = np.zeros(universos.size)

for indice, universo in enumerate(universos):
    numero_personajes_por_universo[indice] = personajes[personajes['Universe'] == universo].Fighter.count()

plot.figure(2)
plot.bar(universos, numero_personajes_por_universo)
plot.xticks(rotation=90)
plot.xlabel('Universos')
plot.ylabel('No. de personajes por universo')
plot.title('Personajes por universo')
```

## Número de personajes por género

Los personajes dentro del universo de Smash Bros. es bastante variado, incluyendo personajes que pueden ser escogidos tanto en su versión masculina como femenina. Analizando los datos, se obtuvieron los siguientes resultados:

```python
generos = personajes['Gender'].unique()
porcentaje_personajes_por_genero = np.zeros(generos.size)

for indice, genero in enumerate(generos):
    porcentaje_personajes_por_genero[indice] = (personajes[personajes['Gender'] == genero].Fighter.count()*personajes['Fighter'].count())/100

plot.figure(3)
plot.pie(porcentaje_personajes_por_genero, 
         labels = generos, 
         autopct="%1.2f%%")
plot.title('Porcentaje de personajes por genero')
```

## Apariciones de los personajes

La franquicia de Smash Bros. nace en 1999 con la aparición de personajes muy queridos para Nintendo tales como Mario, Link o Samus de Metroid, con cada entrega del juego, se han ido incluyendo nuevos y variados personajes, no sólo de Nintendo, sino de otras empresas dedicadas al mercado de videojuegos. En el 2018 se lanzó la última versión del juego, luego de casi 20 del lanzamiento del juego original y con esto, se han incluido un nuevo número de personajes. Al analizar a los personajes más antigüos, es decir, los que se han mantenido desde el juego original y los personajes del último lanzamiento, se obtuvo los siguientes resultados:

### Lista de personajes antigüos

Se define una columna 'Count' en la cual se almacenan los datos de los personajes por juego a través de la siguiente variable:

```python
personajes['Count'] = personajes.count(axis=1) - 4 # Se resta 3 que corresponden a las columnas 'Fighter', 'Universe' y 'Gender'
```

Se resta a 4 la variable anterior para descartar las columnas 'Fighter', 'Universe', 'Gender' y la columna generada 'Count'. Con esto, se obtiene la lista de los personajes antigüos:

```python
personajes_antiguos = personajes.sort_values(ascending = False, by = 'Count').query('Count == 5')
print(personajes_antiguos.loc[:, ['Fighter']])
```

### Lista de personajes nuevos

De manera similar al proceso anterior, utilizando la misma columna 'Count', se obtienen los datos de los personajes nuevos, incluidos en el juego más reciente:

```python
personajes_antiguos = personajes.sort_values(ascending = False, by = 'Count').query('Count == 1')
print(personajes_antiguos.loc[:, ['Fighter']])
```

## Relación entre la ofensiva y defensiva por cada personaje

```python
plot.figure(figsize = (15, 5))
plot.xticks(rotation = 90)
sns.barplot(x = 'Characters', y = 'Defense', hue = 'Offense', data = personajes_SSB4.groupby(['Characters', 'Offense']).sum().reset_index()).set_title('Defensiva vs Ofensiva de los personajes')
```

## Velocidad de caída por personajes

```python
velocidad_caida_personaje = (personajes_SSB4['Fall Speed'].groupby(personajes_SSB4.Characters)).sum()
velocidad_caida_personaje.plot(figsize = (10,5), linewidth = 1, color = 'red')
plot.xlabel('Character')
plot.ylabel('Character usage')
plot.show()
```
